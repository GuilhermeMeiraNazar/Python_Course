```python
import os
from google.colab import userdata
os.environ["GOOGLE_API_KEY"] = userdata.get('Gemini_API_KEY')
from google import genai
client = genai.Client()
```

```python
lista_de_emails = [
    "Estou entrando em contato para verificar a possibilidade de você, quando tiver um tempo entre uma tarefa e outra, enviar aquele relatório atualizado que comentamos rapidamente na última reunião. Sei que é algo simples, mas como preciso alinhar com outras pessoas, seria ótimo receber o quanto antes para manter tudo organizado. Agradeço desde já pela atenção.",

    "Queria confirmar se há a possibilidade de fazermos um pequeno ajuste no horário da nossa conversa de amanhã, pois surgiu um compromisso inesperado que acabou se sobrepondo e, para evitar confusão, achei melhor avisar com antecedência. Se puder me retornar, fico agradecido.",

]
```

```python
def resumir_Emails(emails):
    prompt = ""

    for n in range(len(emails)):
        prompt = prompt + ("Email " + str(n) + ": " + emails[n] + "\n")

    resposta = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=(
            "Chat, faça um resumo do email ou da lista de emails. "
            "Cada email deve ter no máximo 1 linha, respeitando as normas ABNT.\n\n"
            + prompt
        )
    )

    print(resposta.candidates[0].content.parts[0].text)


```

```python
resumir_Emails(lista_de_emails)
```
