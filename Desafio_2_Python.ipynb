```python
import pandas as pd
```

```python
import os
from groq import Groq
from google.colab import userdata
api_key = userdata.get('Groq_API_KEY')
client = Groq(api_key=api_key)
```

/////Lista de perguntas

```python
lista_de_perguntas = [
    "Qual é o nome do protagonista da série de jogos 'The Legend of Zelda'?",
    "Em que ano foi lançado o primeiro console PlayStation?",
    "Qual é o jogo mais vendido de todos os tempos até hoje?",
    "Qual empresa é a criadora da franquia 'Super Mario Bros.'?",
    "No jogo 'Minecraft', qual é o nome do inimigo que explode ao chegar perto do jogador?",
    "Qual é o nome do reino onde se passa a maior parte dos jogos de 'Elden Ring'?",
    "Qual foi o primeiro jogo eletrônico a apresentar o personagem Donkey Kong?",
    "Na franquia 'Halo', qual é o nome da inteligência artificial que acompanha o Master Chief?",
    "Qual é o gênero de jogo conhecido pela sigla MOBA?",
    "Em 'The Witcher 3', qual é o apelido do protagonista Geralt de Rívia?"
]
```

```python
with open("Python_desafio_2.txt", "w", encoding="utf-8") as arquivo:
    for n in lista_de_perguntas:
        arquivo.write(f"{n}\n")
```

```python
lista_desafio_2 = []
with open("Python_desafio_2.txt", "r", encoding="utf-8") as arquivo:
    for n in arquivo:
        lista_desafio_2.append(n)
```

///// Funcao Groq

```python
def LLM(pergunta):
    completion = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[{"role": "user", "content": pergunta}],
        stream=False,
    )
    return completion.choices[0].message.content
```

/////

```python
import csv
with open("respostas.csv", "a", encoding="utf-8") as arquivo_csv:
    for n, pergunta in enumerate(lista_desafio_2):
        resposta_ia = LLM(pergunta)
        resposta_limpa = resposta_ia.replace("\n", " ").replace(",", ";")
        pergunta_limpa = pergunta.replace("\n", " ").replace(",", ";")
        arquivo_csv.write(f"{n},{pergunta_limpa},{resposta_limpa}\n")

        print(f"Pergunta {n} processada com sucesso!")
```
